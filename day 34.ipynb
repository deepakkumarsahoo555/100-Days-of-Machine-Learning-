### Working with Dates

```python
# Converting to datetime datatype
date['date'] = pd.to_datetime(date['date'])
date.info()
```

### 1. Extract year

```python
date['date_year'] = date['date'].dt.year

date.sample(5)
```

## 2. Extract Month

```python
date['date_month_no'] = date['date'].dt.month

date.head()
```

```python
date['date_month_no'] = date['date'].dt.month

date.head()
```

## Extract Days

```python
date['date_day'] = date['date'].dt.day

date.head()
```

```python
# day of week
date['date_dow'] = date['date'].dt.dayofweek

date.head()
```

```python

# day of week - name

date['date_dow_name'] = date['date'].dt.day_name()

date.drop(columns=['product_id','city_id','orders']).head()
```

```python
# is weekend?

date['date_is_weekend'] = np.where(date['date_dow_name'].isin(['Sunday', 'Saturday']), 1,0)

date.drop(columns=['product_id','city_id','orders']).head()
```

## Extract week of the year

```python
date['date_week'] = date['date'].dt.week

date.drop(columns=['product_id','city_id','orders']).head()
```

```python
# Use isocalendar().week instead of .week which is deprecated
date['date_week'] = date['date'].dt.isocalendar().week

date.drop(columns=['product_id','city_id','orders']).head()
```

## Extract Quarter

```python

date['quarter'] = date['date'].dt.quarter

date.drop(columns=['product_id','city_id','orders']).head()
```

## Extract Semester

```python
date['semester'] = np.where(date['quarter'].isin([1,2]), 1, 2)

date.drop(columns=['product_id','city_id','orders']).head()
```

## Extract Time elapsed between dates

```python
import datetime

today = datetime.datetime.today()

today
```

```python
today - date['date']
```

```python
(today - date['date']).dt.days
```

```python
# Months passed
# Using 'D' (days) instead of 'M' and converting to approximate months
# A month is approximated as 30.44 days (365.25/12)
np.round((today - date['date']) / np.timedelta64(1, 'D') / 30.44, 0)

# Alternative approach using pandas if available:
# import pandas as pd
# pd.to_datetime(today).to_period('M') - pd.to_datetime(date['date']).to_period('M')
```
